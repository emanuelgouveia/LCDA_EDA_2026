```python
import pandas as pd
import numpy as np

df=pd.read_csv("LED.csv")

print("O numero de observações é:", len(df.index))
print("O numero de colunas é:", len(df.columns))
```

```python
print(df.dtypes)
for col in df.columns:
    if pd.api.types.is_string_dtype(df[col]):
        print("A variável", col, "é discreta qualitativa")
    elif df[col].dtype == int:
        print("A variável", col, "é discreta")
    else: print("A variável", col, "é continua")
```

```python
print(df['Country'].value_counts())
print(df['Status'].value_counts())
```

```python
print(pd.isna(df).sum())
print(df.dropna().count())
```

```python

df[df.duplicated(subset=['Country', 'Year'], keep=False)]
```

```python
grupo = df[(df["Status"] == "Developing") & (df["Year"] == 2007)]

baixa_life_exp = grupo.sort_values(by="Life expectancy ",ascending=True)[["Country", "Life expectancy "]].head(10)

alta_life_exp = grupo.sort_values(by="Life expectancy ",ascending=False)[["Country", "Life expectancy "]].head(10)

print("Os 10 países em desenvolvimento com esperança de vida mais baixa em 2007 foram:")
print(baixa_life_exp)

print("Os 10 países em desenvolvimento com esperança de vida mais alta em 2007 foram:")
print(alta_life_exp)


```

```python
mediagrupo=grupo.mean(numeric_only=True)
mediatudo=df.mean(numeric_only=True)

print("A média de cada coluna nos paises em desenvolvimento em 2007 foram:", mediagrupo)
print("A média geral de cada coluna (todos os paises, todos os anos) foram:",mediatudo )

```
