```python
total_duplicados = df.duplicated().sum()
print(f"Total de linhas duplicadas: {total_duplicados}")

total_duplicados = df.duplicated(["Suburb","Address","Postcode"]).sum()
print(f"Total de casas duplicadas: {total_duplicados}")

total_duplicados = df.duplicated(["Suburb","Address","Postcode","Rooms"]).sum()
print(f"Total de casas duplicadas com o mesmo numero de quartos: {total_duplicados}")

print("Total de casas com a mesma morada, suburbio e o mesmo numero de casa mas com diferente número de quartos: 112")
```

```python
# apenas casas com 4 quartos e print(f"A média de preço para casas com 4 quartos é: {media_4_quartos:,.2f}")media 4 quartos
grupo = df[df["Rooms"] == 4]
media_4_quartos = grupo["Price"].mean()
print(f"A média de preço para casas com 4 quartos é: {media_4_quartos:,.2f}")

grupo_interessante = df[(df["Rooms"] == 4) & (df["Price"] < 600000)]

print(f"Encontrámos {len(grupo_interessante)} casas com 4 quartos abaixo de 600 mil.")
```

```python
#Estatísticas para o Dataset
media_preco_total = df["Price"].mean()
media_quartos_total = df["Rooms"].mean()
media_date_total = df["Date"].mean

#Estatísticas para o Grupo (>4 quartos e <600k)
media_preco_grupo = grupo_interessante["Price"].mean()
media_quartos_grupo = grupo_interessante["Rooms"].mean()
media_date_grupo = grupo_interessante["Date"].mean

# comparação
print("--- COMPARAÇÃO DE ESTATÍSTICAS ---")
print(f"Preço Médio: Total = {media_preco_total:,.2f} | Grupo = {media_preco_grupo:,.2f}")
print(f"Média de Quartos: Total = {media_quartos_total:.2f} | Grupo = {media_quartos_grupo:.2f}")
```

```python
date_minimo = grupo_interessante["Date"].min()
date_maximo = grupo_interessante["Date"].max()

data_mais_baixa = grupo_interessante[grupo_interessante["Date"] == date_minimo]
data_mais_alta = grupo_interessante[grupo_interessante["Date"] == date_maximo]

print("--- CLASSIFICAÇÃO MAIS BAIXA (Menor Data) ---")
print(data_mais_baixa[["Address", "Price", "Date"]])

print("\n--- CLASSIFICAÇÃO MAIS ALTA (Maior Data no Grupo) ---")
print(data_mais_alta[["Address", "Price", "Date"]])
```
