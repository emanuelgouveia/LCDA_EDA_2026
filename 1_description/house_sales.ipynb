2 - Identificar e documentar os tipos das variáveis(Discretas/contínuas)

Caso sejam discretas são nominais ou ordinais, caso não o sejam a diferença intervalar é relevante?

Dataset de teste

```python
for column in df1.columns:
    if df1[column].dtype == 'str':
        print(f'Coluna: {column} - Tipo: Discreta - Subtipo: Nominal')
    elif df1[column].dtype == 'int64':
        print(f'Coluna: {column} - Tipo: Discreta - Subtipo: Ordinal')
    else:
        print(f'Coluna: {column} - Tipo: Contínua')
```

Número de missing values nos datasets

Dataset de teste

```python
for column in df1.columns:
    if df1[column].isnull().sum() > 0:
        num_missing = df1[column].isnull().sum()
        print(f'Coluna: {column} - Número de missing values: {num_missing}')

total_missing1 = df1.isnull().sum().sum()
print(f'Número total de missing values no dataset de teste: {total_missing1}')
```

Duplicate observations in dataset yes/no

dataset teste

```python
while True:
    for column in df1.columns:
        num_duplicates = df1.duplicated(subset=column).sum()
        if num_duplicates > 0:
            print(f'Coluna: {column} - Número de observações duplicadas: {num_duplicates}')
    break
```
